```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import LogisticRegression
```

## Question 1

```python
q1 = pd.read_excel("Real estate valuation data set.xlsx")
```

```python
q1.info()
```

```python
q1.insert(0,"X0",1)
q1
```

```python
for i in [80,60,70,90]:
    q1_train = q1.iloc[:(len(q1)*i)//100]
    q1_test = q1.iloc[len(q1)-(len(q1)*i)//100:]
    
    print('\033[1m'+f"For {i}:{100-i} Split"+ '\033[0m')

    y_1_train = q1_train["Y house price of unit area"]
    X_train = q1_train[["X0","X1 transaction date","X2 house age","X3 distance to the nearest MRT station","X4 number of convenience stores","X5 latitude","X6 longitude"]]
    X_train = X_train.values
    X_train = np.asmatrix(X_train)
    y_1_train = y_1_train.values

    x_xt_inv = np.linalg.inv(np.matmul(X_train.transpose(),X_train))
    beta_1 = np.matmul(np.matmul(x_xt_inv,X_train.transpose()),y_1_train)

    for i in range(7):
        if i == 0:
            print("Intercept: ",beta_1[0,0])
        else:
            print(f"Beta {i}: ",beta_1[0,i])

    X_test = q1_test[["X0","X1 transaction date","X2 house age","X3 distance to the nearest MRT station","X4 number of convenience stores","X5 latitude","X6 longitude"]]
    X_test = X_test.values
    y_1_pred = np.matmul(beta_1,X_test.transpose())
    y_1_test = q1_test["Y house price of unit area"]
    y_1_test = np.matrix(y_1_test.values)
    y_1_err = np.subtract(y_1_test,y_1_pred)
    error_1 = np.dot(y_1_err,y_1_err.transpose())
    error_1 = error_1[0,0]
    print(f"Mean Squared Error for Y (House Price of unit area) : {error_1}")

    rss = error_1
    tss_vec = y_1_test - np.mean(y_1_test)
    tss = np.dot(tss_vec,tss_vec.transpose())
    tss = tss[0,0]
    r_square = 1 - (rss/tss)
    print(f"Coefficient of Determination (r^2): {r_square}\n")
```

### Ridge Model:

```python
for alpha in [0.001, 0.005, 0.01, 0.05, 0.1]:
    print('\033[1m'+f"\nALPHA = {alpha} \n"+ '\033[0m')
    for i in [80,60,70,90]:
        q1_train = q1.iloc[:(len(q1)*i)//100]
        q1_test = q1.iloc[len(q1)-(len(q1)*i)//100:]
    
        print('\033[1m'+f"For {i}:{100-i} Split"+ '\033[0m')

        y_1_train = q1_train["Y house price of unit area"]
        X_train = q1_train[["X1 transaction date","X2 house age","X3 distance to the nearest MRT station","X4 number of convenience stores","X5 latitude","X6 longitude"]]
        X_test = q1_test[["X1 transaction date","X2 house age","X3 distance to the nearest MRT station","X4 number of convenience stores","X5 latitude","X6 longitude"]]
        y_1_test = q1_test["Y house price of unit area"]

        rr = Ridge(alpha=alpha)
        rr.fit(X_train,y_1_train)
    
        Ridge_test_score = rr.score(X_test, y_1_test)
    
        y_1_pred = rr.predict(X_test)
        y_1_test = np.matrix(y_1_test.values)
        y_1_err = np.subtract(y_1_test,y_1_pred)
        error_1 = np.dot(y_1_err,y_1_err.transpose())
        error_1 = error_1[0,0]
        print(f"Mean Squared Error for Y (House Price of unit area) : {error_1}")
        print(f"Coefficient of Determination (r^2): {Ridge_test_score}")
    
```

```python
q2 = pd.read_csv("haberman.data",delimiter=",",header = None)
```

```python
q2 = q2.set_axis(['X1', 'X2', 'X3', 'Y'], axis=1, inplace=False)
```

```python
q2 
```

```python
lr = LogisticRegression()
for i in [80,60,70,90]:
    q2_train = q2.iloc[:(len(q2)*i)//100]
    q2_test = q2.iloc[len(q2)-(len(q2)*i)//100:]
    
    print('\033[1m'+f"For {i}:{100-i} Split"+ '\033[0m')

    y_2_train = q2_train["Y"]
    X_train = q2_train[["X1","X2","X3"]]
    X_test = q2_test[["X1","X2","X3"]]
    y_2_test = q2_test["Y"]
    
    lr = LogisticRegression()
    lr.fit(X_train,y_2_train)
    coefficients = lr.coef_
    intercept = lr.intercept_
    
    print(f"Intercept = {intercept[0]}")
    for j in range(3):
        print(f"Beta {j+1} = {coefficients[0,j]}")
    
    print("\nPredicted Survival Status: ",lr.predict(X_test),"\n")
    
    y_2_pred = lr.predict(X_test)
    y_2_test = y_2_test.values
    sum = 0
    
    diff = np.subtract(y_2_pred,y_2_test)
    diff = list(diff)
    
    for element in diff:
        if element != 0:
            sum += 1
    error = sum/len(y_2_test)
    
    print(f"Classification Error : {error} \n")
```
